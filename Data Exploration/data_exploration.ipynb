## Data Exploration ##

```python
# Import libraries
import duckdb
import pandas as pd
import os
```

```python
# Load Data
db = duckdb.connect(database='../Data/jaffle_shop.duckdb')
```

```python
db.execute("SHOW TABLES").df()
```

```python
db.execute("DESCRIBE customers").df()
```

```python
db.execute("DESCRIBE order_items").df()
```

```python
db.execute("DESCRIBE orders").df()
```

```python
db.execute("DESCRIBE products").df()
```

```python
# Lenght of each table
table_lengths = pd.DataFrame({
    "table_name": ["customers", "orders", "order_items", "products"],
    "length": [
        db.execute("SELECT COUNT(*) FROM customers").fetchone()[0],
        db.execute("SELECT COUNT(*) FROM orders").fetchone()[0],
        db.execute("SELECT COUNT(*) FROM order_items").fetchone()[0],
        db.execute("SELECT COUNT(*) FROM products").fetchone()[0]
    ]
})
table_lengths
```

```python
# Explore first 5 rows of each table
db.execute("SELECT * FROM customers LIMIT 5").df()
```
